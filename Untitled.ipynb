```python
from quantrocket import get_prices
prices = get_prices("usstock-free-1d",universes="usstock-aapl", start_date="01-01-2023",end_date = "12-31-2023", fields=["Close"])
print(prices)
print(prices.columns)
```

```python
prices_list = prices['FIBBG000B9XRY4'].tolist()
print(prices_list)
```

```python
rd = []
for i in range(1, len(prices_list)):
        if i > 1:
            #If rounding off needed to 2 decimal points we could use this.
            #rd.append(round((prices_list[i] - prices_list[i-1]) / (prices_list[i-1]),2))
            rd.append((prices_list[i] - prices_list[i-1]) / (prices_list[i-1]))
print(rd)
print(len(rd))
```

```python
sd = []
for i in range(0, len(rd)):
    if rd[i] >= 0.01:
        sd.append(1)
    elif rd[i] > -0.01:
        sd.append(0)
    else:
        sd.append(-1)
print(sd)
```
